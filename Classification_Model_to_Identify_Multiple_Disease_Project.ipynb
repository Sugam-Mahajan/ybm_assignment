# **‚≠ê Classification Model to Identify Multiple Disease**

```python
# import library
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# import data
disease = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/MultipleDiseasePrediction.csv')
```

```python
# view data
disease.tail()
```

```python
# info of data
disease.info()
```

```python
# check for missing value
disease.isnull().sum()
```

```python
# check for categories
disease.nunique()
```

```python
# correlation
disease.corr()
```

```python
# visualize pairplot
sns.pairplot(disease)
```

```python
# column names
disease.columns()
```

```python
# define y
y= disease['prognosis']
```

```python
# define X
x=disease.drop('prognosis',axis=1)
```

```python
# split data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.7,random_state=2529)
```

```python
# verify shape
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# select model
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier()
```

```python
# train model
model.fit(x_train,y_train)
```

```python
# predict with model
y_pred=model.predict(x_test)
```

```python
# model evaluation
```

```python
# model accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# model confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# model classification report
from sklearn.metrics import precision_recall_fscore_support
report=precision_recall_fscore_support(y_test,y_pred)
print((report))
```
